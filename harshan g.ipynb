```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
data=pd.read_csv('C:/Users/ghars/Downloads/titanic_dataset.csv')
data
```

```python
# Set 'PassengerId' as the index column
data.set_index('PassengerId', inplace=True)
data
```

```python
# Fill missing values with the mean value for each column
data = data.fillna(data.mean())
data
```

```python

# Fill missing values in the 'Embarked' column with the most common value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

data
```

```python

# You could also fill missing values with the mode value for each column
data = data.fillna(data.mode().iloc[0])
```

```python
data
```

```python
# Use the 'boxplot' method to check for outliers in the 'Age' column
data.boxplot(column='Age')
plt.show()
```

```python
from scipy import stats

# Use the 'zscore' function to calculate the z-scores of the 'Fare' column
z_scores = stats.zscore(data['Fare'])



z_scores
```

```python
 # Keep only the rows that have z-scores between -3 and 3
data = data[(z_scores > -3) & (z_scores < 3)]
```

```python
data
```

```python

# Use the 'quantile' method to find the 25th and 75th percentile values of the 'Pclass' column
p25 = data['Pclass'].quantile(0.25)
p75 = data['Pclass'].quantile(0.75)

```

```python
p25
```

```python
p75
```

```python
# Keep only the rows that have 'Pclass' values between the 25th and 75th percentiles
data = data[(data['Pclass'] >= p25) & (data['Pclass'] <= p75)]
```

```python
data
```

```python
# Visualize the distribution of 'Age' to identify outliers
sns.boxplot(data['Age'])
plt.show()
```

```python
# Remove outliers using the interquartile range method
q1 = data['Age'].quantile(0.25)
q3 = data['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
data = data[(data['Age'] > lower_bound) & (data['Age'] < upper_bound)]
data
```

```python
# Repeat the process for 'Fare' and 'Pclass'
sns.boxplot(data['Fare'])
q1 = data['Fare'].quantile(0.25)
q3 = data['Fare'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
data = data[(data['Fare'] > lower_bound) & (data['Fare'] < upper_bound)]
```

```python

sns.boxplot(data['Pclass'])
q1 = data['Pclass'].quantile(0.25)
q3 = data['Pclass'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
data = data[(data['Pclass'] > lower_bound) & (data['Pclass'] < upper_bound)]
```
